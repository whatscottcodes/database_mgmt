```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('PaceDashboard.db')
```

```python
c = conn.cursor()
```

```python
tables = [table_info[0] for table_info in c.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()]
```

```python
cols_dict = {}

for table in tables:
    print(table)
    cols_dict[table] = pd.read_sql(f"SELECT * FROM {table} LIMIT 5", conn).columns.to_list()
    print('--------------------------------')
```

```python
conn.close()
```

### This gives us a DF with the current database's columns by table, should add the columns of the dataframes that go into the database. 

```python
pd.DataFrame.from_dict(cols_dict, orient='index')#.to_csv('col_names_raw.csv')
```
